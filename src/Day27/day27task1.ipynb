```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("C:/Users/redmi/Downloads/canada_per_capita_income.csv")

print(df.head())

# Prepare data
X = df[['year']]          # Independent variable
y = df['per capita income (US$)']   # Dependent variable

# Create model
model = LinearRegression()
model.fit(X, y)

# Predict for 2020
income_2020 = model.predict([[2020]])
print("Predicted income for 2020:", income_2020[0])

# Predict from 2020 to 2030
years = np.arange(2020, 2031).reshape(-1, 1)
future_predictions = model.predict(years)

# Print predictions
for year, income in zip(range(2020, 2031), future_predictions):
    print(f"Year {year} → Predicted Income: {income}")

# Plot graph
plt.scatter(X, y, color='blue')
plt.plot(years, future_predictions, color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```
